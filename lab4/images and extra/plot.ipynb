```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
filename = ['test/process1.dat.sjf.csv', 'test/process2.dat.sjf.csv', 'test/process3.dat.sjf.csv', 'test/process1.dat.RR.csv', 'test/process2.dat.RR.csv', 'test/process3.dat.RR.csv']
keys = ["Throughput", "Average Repsonse Time", "Average Waiting Time", "Average TurnAround Time", "Average Penalty Ratio"]
data ={"Average Repsonse Time" : {"RR" : [], "SJF" : []}, "Average Waiting Time" : {"RR" : [], "SJF" : []}, "Average TurnAround Time" : {"RR" : [], "SJF" : []}, "Average Penalty Ratio" : {"RR" : [], "SJF" : []}}
```

```python
data ={"Average Turnaround Time" : {"RR" : [871.571, 248.444, 1225.83], "SJF" : [754.286, 214.5, 911.833]}, 
       "Average Waiting Time" : {"RR" : [657.714, 188.278, 1022.08], "SJF" : [544.286, 157.5, 717.417]},  
       "Average Penalty Ratio" : {"RR" : [3.9886, 2.82998, 5.14117], "SJF" : [4.81304, 2.51155, 3.69326]},}
```

```python
for key in data:
    X = ['Test 1','Test 2','Test 3']
    SJF = data[key]['SJF']
    RR = data[key]['RR']
    plt.figure(figsize=(10, 7))
    X_axis = np.arange(len(X))
    
    plt.bar(X_axis - 0.2, SJF, 0.4, label = 'SJF', color='pink')
    plt.bar(X_axis + 0.2, RR, 0.4, label = 'RR', color='grey')
    
    plt.xticks(X_axis, X)
    plt.xlabel("Test Data No.")
    plt.ylabel(key)
    plt.title("Scheme vs "+key)
    plt.legend()
    plt.savefig("output_" + key + ".png")
    plt.show()
```
